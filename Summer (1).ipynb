### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
import pandas as pd 
df = pd.read_csv("summer.csv",dtype = object)
df = df[df['Medal']=='Gold']
data =[]
for sport in df['Sport'].unique():
    data.append([sport,len(df[df['Sport']==sport])])
Data = pd.DataFrame(data,columns=['Sport','Gold Medal']).sort_values(by = 'Gold Medal',ascending = False).head()
Data.plot(kind='bar',color='purple',x='Sport')
Data.head()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
import pandas as pd 
df = pd.read_csv("summer.csv",dtype = object)
data =[]
for sport in df['Sport'].unique():
    data.append([sport,len(df[df['Sport']==sport])])
Data = pd.DataFrame(data,columns=['Sport','Medal']).sort_values(by = 'Medal',ascending = False).head()
Data.plot(kind='bar',color='green',x='Sport',y= 'Medal')
Data.head()
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
# ans = 1928

import pandas as pd
df = pd.read_csv("summer.csv")
df = df[df['Country']=='IND']
df = df[df['Medal']=='Gold']

data =[]
for year in df['Year'].unique():
    data.append([year,len(df[df['Year']==year])])
pd.DataFrame(data,columns=['Year','Gold Medal']).sort_values(by = 'Year',ascending = True).head(1)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
import pandas as pd
df = pd.read_csv("summer.csv")
data =[]
for event in df['Event'].unique():
    data.append([event,len(df[df['Event']==event])])
Data = pd.DataFrame(data,columns=['Event','Athlete']).sort_values(by = 'Athlete',ascending = False).head()
Data.plot(kind='bar',color='pink',x='Event')
Data.head()
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
import pandas as pd
df = pd.read_csv("summer.csv",dtype = object)
df = df[df['Gender']=='Women']
df = df[df['Medal']=='Gold']

data =[]
for sport in df['Sport'].unique():
    data.append([sport,len(df[df['Sport']==sport])])
Data = pd.DataFrame(data,columns=['Sport','Womens']).sort_values(by = 'Womens',ascending = False).head()
Data.plot(kind='bar',color='red',x='Sport')
Data.head()
```
